```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Play Store Data.csv")

print("Data set loaded successfully...")
```

```python
df.info()
```

```python
df['Reviews']=pd.to_numeric(df['Reviews'],errors='coerce')
```

```python
df['Installs']=df["Installs"].str.replace('+','',regex=False).str.replace(',','',regex=False)
df['Installs']=pd.to_numeric(df['Installs'],errors='coerce')
```

```python
def size_to_mb(x):
    if 'M' in str(x):
        return float(str(x).replace("M",''))
    elif 'k' in str(x):
        return float(str(x).replace('k',''))/1024
    else:
        return None
df["Size_MB"] = df["Size"].apply(size_to_mb)
df["Last_Updated_dt"] = pd.to_datetime(df["Last Updated"], errors = "coerce")
```

```python
df.head()
```

```python
filtered = df[
    (df['Rating']>=4.0) &
    (df['Size_MB']>=10) &
    (df['Last_Updated_dt'].dt.month == 1)
]
filtered
```

```python
installs_by_category = filtered.groupby('Category')['Installs'].sum()
installs_sorted = installs_by_category.sort_values(ascending=False)
top_10_categories = installs_sorted.head(10)
print("The top 10 categories by installs are: ")
print(top_10_categories)
```

```python
top_10_data = filtered[ filtered['Category'].isin(top_10_categories.index) ]
avg_rating = top_10_data.groupby('Category')['Rating'].mean()
total_reviews = top_10_data.groupby('Category')['Reviews'].sum()

data = pd.DataFrame({
    'average_rating': avg_rating,
    'total_reviews': total_reviews
})

print("Aggregated data for top 10 categories:")
print(data)
```

```python
categories = list(data.index)
x = np.arange(len(categories))
width = 0.35
plt.figure(figsize=(10,6))
plt.bar(x - width/2, data['average_rating'], width, color = 'blue', label = 'Average Rating')
plt.bar(x + width/2, data['total_reviews']/1000000, width, color = 'orange', label = 'Total Reviews(millions)')

plt.xticks(x, categories, rotation=25)
plt.xlabel("App Categories")
plt.ylabel("values")
plt.title("Top 10 categories - Average Rating vs Total Reviews")
plt.show()
    
```
